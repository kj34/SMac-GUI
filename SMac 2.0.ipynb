```python
from tkinter import *
from tkinter import filedialog
from PIL import Image, ImageTk
import re
import os

root = Tk()
root.geometry("1000x500") 
root.title("SMac V2.0")

file_path_string = ""
bjdresult = ""
benzene_result = ""
bs2_result = ""


def Solvent(fin, file_data):
    try:
        if "benzene" in fin.name:
            benzene_result = re.search("SCF Done:\s+\S+\s=\s*(-?[0-9.]+)", file_data)
            a = Label(root, text = fin.name)
            b = Label(root, text = "SOLV VALUE:  " + benzene_result.group(1))
            a.pack()
            b.pack()
    except:
        pass

def Bjd(fin, file_data):
    try:
        if "bjd" in fin.name:
            bjdresult = re.search("SCF Done:\s+\S+\s=\s*(-?[0-9.]+)", file_data)
            a = Label(root, text = fin.name)
            b = Label(root, text = "BJD3 VALUE:  " + bjdresult.group(1))
            a.pack()
            b.pack()
            
    except:
        pass
            
def Bs2(fin, file_data):
    try:
        if "bs2" in fin.name:
            bs2_result = re.search("SCF Done:\s+\S+\s=\s*(-?[0-9.]+)", file_data)
            a = Label(root, text = fin.name)
            b = Label(root, text = "BS2 VALUE:  " + bs2_result.group(1))
            a.pack()
            b.pack()
    except:
        pass
        
        
def Parent(fin, file_data):      
    try:
        if "bjd" not in fin.name and "benzene" not in fin.name and "bs2" not in fin.name:
            parent_result = list(re.findall("SCF Done:\s+\S+\s=\s*(-?[0-9.]*)", file_data))
            scf_parent = parent_result[-1]
            a = Label(root, text = fin.name)
            b = Label(root, text = "SCF VALUE:  " + scf_parent)
            #collected_data["scf_parent_dict"].append(scf_parent)
            zero_point_energies = re.search("zero-point Energies=\s+(-?[0-9.]+)", file_data)
            c = Label(root, text = "H VALUE:  " + zero_point_energies.group(1))
            #collected_data["zero_point_dict"].append(zero_point_energies.group(1))
            thermal_free_energies = re.search("thermal Free Energies=\s+(-?[0-9.]+)", file_data)
            d = Label(root, text = "G VALUE:  " + thermal_free_energies.group(1))
            #collected_data["free_energy_dict"].append(thermal_free_energies.group(1))
            a.pack()
            b.pack()
            c.pack()
            d.pack()
    except:
        pass


def get_file():
    with open (file_path_string.name, "r") as fin:
        file_data = fin.read()
        Bjd(fin, file_data)
        Solvent(fin, file_data)
        Bs2(fin, file_data)
        Parent(fin, file_data)
        
def browse_file():
    global file_path_string
    file_path_string = filedialog.askopenfile()  
    get_file()
    

path = "SMac logo.png"
img = ImageTk.PhotoImage(Image.open(path))
panel = Label(root, image = img)
panel.pack()   

browse_file = Button(root, text = "Browse for File", padx = 40, pady = 20, command = browse_file)
browse_file.pack()    

mainloop()
```
